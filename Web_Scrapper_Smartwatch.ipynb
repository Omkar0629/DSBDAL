```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link = "https://www.flipkart.com/noise-evolve-3-bt-calling-1-43-inch-amoled-always-on-display-metallic-design-smartwatch/product-reviews/itm36b6117652964?pid=SMWGGP58M2H75R6H&lid=LSTSMWGGP58M2H75R6HQDJ44Y&marketplace=FLIPKART"
page=requests.get(link)
page
```

```python
soup=bs(page.content,'html.parser')
print(soup.prettify())
```

```python
names=soup.find_all('p',{'class':'_2sc7ZR _2V5EHH'})
names
```

```python
cname=[]
for i in range(0,len(names)):
    cname.append(names[i].get_text())
cname=cname[0:10]
cname
```

```python
title=soup.find_all('p',{'class':'_2-N8zT'})
title
```

```python
rtitle=[]
for i in range(0,len(title)):
    rtitle.append(title[i].get_text())
rtitle=rtitle[0:10]
rtitle
```

```python
review=soup.find_all('div',{'class':'t-ZTKy'})
review
```

```python
rreview=[]
for i in range(0,len(review)):
    rreview.append(review[i].get_text())
rreview=rreview[0:10]
rreview
```

```python
rating=soup.find_all('div',{'class':'_3LWZlK _1BLPMq'})
rating
```

```python
rrating=[]
for i in range(0,len(rating)):
    rrating.append(rating[i].get_text())
rrating=rrating[0:10]
rrating
```

```python
data={
    "name":cname,
    "title":rtitle,
    "rating":rrating,
    "review":rreview
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_csv('smartwatch.csv')
```
